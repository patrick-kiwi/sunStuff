```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
# Planet masses (in solar masses)
m_jupiter = 0.0009546
m_saturn = 0.0002857
m_uranus = 0.00004366
m_neptune = 0.00005149

# Orbital radii (in AU)
a_jupiter = 5.20
a_saturn = 9.58
a_uranus = 19.18
a_neptune = 30.07

# Orbital periods (in years)
T_jupiter = 11.86
T_saturn = 29.46
T_uranus = 84.01
T_neptune = 164.8

# Amplitudes (in AU)
A_jupiter = m_jupiter * a_jupiter
A_saturn = m_saturn * a_saturn
A_uranus = m_uranus * a_uranus
A_neptune = m_neptune * a_neptune

# Time array (0 to 200 years, 1000 points)
t = np.linspace(0, 20000, 1000)

# Sun's x-displacement due to each planet (cosine functions)
x_jupiter = A_jupiter * np.cos(2 * np.pi * t / T_jupiter)
x_saturn = A_saturn * np.cos(2 * np.pi * t / T_saturn)
x_uranus = A_uranus * np.cos(2 * np.pi * t / T_uranus)
x_neptune = A_neptune * np.cos(2 * np.pi * t / T_neptune)

# Total x-displacement
x_sun = x_jupiter + x_saturn + x_uranus + x_neptune

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(t, x_sun, label='Sun\'s x-displacement', color='b')
plt.title('Sun\'s Motion Around Solar System Barycenter (x-component)')
plt.xlabel('Time (years)')
plt.ylabel('Displacement (AU)')
plt.grid(True)
plt.legend()
plt.savefig('sun_barycenter_motion.png')
```
